### Mergin Time Series

```python
for key, table in tables.items():
    heads = [head for head in table.columns if "DAYS" in head or "MONTHS" in head or "YEARS" in head]
    print(key,":", heads)
```

app_train und app_test sind zeitbezogene Stammdaten und somit keine Zeitreihen

### Bureau

```python
ts = [head for head in bureau.columns if "DAYS" in head or "MONTHS" in head or "YEARS" in head]
ts
```

```python
bureau[["SK_ID_CURR", "SK_ID_BUREAU"] + ts].head()
```

```python
quants = [0, 0.25, 0.5, 0.75, 1]
for index, head in enumerate(ts):
    a = bureau[["SK_ID_CURR", head]]
    a = a.groupby(by = ["SK_ID_CURR"]).quantile(quants)
    a = a.unstack(level=-1)
    if index == 0:
        b = a
        continue
    b = b.join(a , on="SK_ID_CURR")
b.head()
```

```python
# Vereinigungsmenge
app_train = app_train.join(b, on="SK_ID_CURR")
```
