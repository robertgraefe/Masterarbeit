```python
app_train.head()
```

```python
# Application Train & Application Test enthalten die selben Attribute (außer TARGET)
todo = list(app_test.columns)
```

### Keine Datenaufbereitung für Primär- und Fremdschlüssel

```python
keys = ["SK_ID_CURR"]
todo = [element for element in todo if element not in keys]
```

### keine Datenaufbereitung der nominalen Daten

* 1 == True
* 0 == False

```python
app_train[n_heads].head()
```

```python
# Löschen der nominalen Daten von der TODO Liste
todo = [head for head in todo if head not in n_heads]
```

### metrische stetige Daten

* Tage zu Jahre
* Vorzeichenumkehr

```python
app_train[ms_heads].head()
```

```python
days = [head for head in ms_heads if "DAY" in head]

for head in days:
    app_train[head] /= 365
    app_test[head] /= 365
    
    app_train.rename(columns={head : head.replace("DAYS","YEARS")},inplace=True)
    app_test.rename(columns={head : head.replace("DAYS","YEARS")},inplace=True)
```

```python
# Anpassung der Attributbeschreibung
for head in days:
    id = des[des["Row"] == head].index
    des.loc[id,"Row"] = head.replace("DAY", "YEAR")
    des.loc[id,"Description"] = des.loc[id,"Description"].values[0].replace("day", "year")
```

```python
# Löschen der days Daten von der TODO Liste
todo = [head for head in todo if head not in ms_heads]
```

### keine Datenaufbereitung der metrisch diskreten Daten

```python
app_train[md_heads].head()
```

```python
# Löschen der days Daten von der TODO Liste
todo = [head for head in todo if head not in md_heads]
```
