# Datenaufbereitung

```python
# Schnittmenge zwischen app_train und bureau
# Ergänzt die Variable "TARGET" in dem bureau-Datensatz
cc = pd.merge(cc, app_train[["SK_ID_CURR","TARGET"]] ,on="SK_ID_CURR")

# Austausch der numerischen Kategorien durch sprechende, interpretierbare Variablen
cc["TARGET"].replace(
    {
        0: "Payback",
        1: "Default"
    }, inplace = True
)
```

```python
cc.head()
```

# Datenanalyse

```python
payment = cc[["PA_CC_PAYMENT_30","PA_CC_PAYMENT_60","PA_CC_PAYMENT_90","PA_CC_PAYMENT_120","PA_CC_PAYMENT_150","PA_CC_PAYMENT_180"]].copy()
amount = cc[["PA_CC_BALANCE_30","PA_CC_BALANCE_60","PA_CC_BALANCE_90","PA_CC_BALANCE_120","PA_CC_BALANCE_150","PA_CC_BALANCE_180"]].copy()
days = cc[["PA_CC_DPD_30","PA_CC_DPD_60","PA_CC_DPD_90","PA_CC_DPD_120","PA_CC_DPD_150","PA_CC_DPD_180"]].copy()
```

```python
cc["sigma"] = payment.std(axis=1)
cc["sigmaII"] = amount.std(axis=1)
cc["sigmaIII"] = days.std(axis=1)
```

```python
cc.head()
```

```python
payback = cc[cc["TARGET"] == "Payback"]
default = cc[cc["TARGET"] == "Default"]
```

```python
head = "sigma"

pb = payback[head].copy()
df = default[head].copy()

pb.hist()
df.hist()

plt.legend(labels=["Payback","Default"])
plt.xlabel("Sigma Payments")
plt.ylabel("# Kreditnehmer")
plt.show()
```

```python
head = "sigma"

pb = payback[head].copy()
df = default[head].copy()

sns.kdeplot(pb)
sns.kdeplot(df)

plt.legend(labels=["Payback","Default"])
plt.xlabel("Sigma Payments")
plt.ylabel("Wahrscheinlichkeit")
plt.show()
```
